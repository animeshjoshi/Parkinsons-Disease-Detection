```python
y = df['status']
```

```python
X = df.drop('status', axis=1)
```

```python
X = X.drop('name', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
sigmoid_model = LogisticRegression()
```

```python
sigmoid_model.fit(X_train, y_train)
```

```python
predictions_logreg = sigmoid_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,ConfusionMatrixDisplay
```

```python
cm_logreg = confusion_matrix(y_test,predictions_logreg)
```

```python
logregcm_plot = ConfusionMatrixDisplay(cm_logreg).plot()
```

```python
print(classification_report(y_test, predictions_logreg))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=9)
```

```python
knn_model.fit(X_train, y_train)
```

```python
knn_predictions = knn_model.predict(X_test)
```

```python
cm_knn = confusion_matrix(y_test,knn_predictions)
```

```python
knncm_plot = ConfusionMatrixDisplay(cm_knn).plot()
```

```python
print(classification_report(y_test,knn_predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators = 250)
```

```python
rf_model.fit(X_train,y_train)
```

```python
rf_predictions = rf_model.predict(X_test)
```

```python
cm_rf = confusion_matrix(y_test,rf_predictions)
```

```python
rfcm_plot = ConfusionMatrixDisplay(cm_rf).plot()
```

```python
print(classification_report(y_test, rf_predictions))
```
